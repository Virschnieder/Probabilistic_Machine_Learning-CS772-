```python
def squared_exponential(x1,x2,l):
    ans = (x1-x2)**2
    ans = -ans/(l*l)
    t = np.exp(ans)
   # print(t)
    return t
```

```python
def get_matrix(X,n,l):
    kernel = np.zeros((n,n))
    for i in range(n):
        for j in range(n):
            kernel[i][j] = squared_exponential(X[i], X[j],l)
    return kernel
            
```

```python
def gen_data(n,sigma):
    X = np.linspace(0,4*np.pi, num = n)
    Y = np.sin(X)
    for i in range(0,n):
        Y[i] = Y[i] + np.random.normal(0,sigma,1)
    return X,Y
```

```python
X,Y = gen_data(100, np.sqrt(0.05))
```

```python
l = [0.2,0.5,1,2,10]
import random
```

```python
import matplotlib.pyplot as plt
```

```python
for num in l:
    mean = np.zeros(100)
    K = get_matrix(X,100,num)
    f = np.random.multivariate_normal(mean,K)
    mean = np.dot(np.linalg.inv(np.linalg.inv(K)*0.05 + np.eye(100)),Y)
    plt.plot(X,f)
    plt.plot(X,mean)
    plt.plot(X, np.sin(X))
    ax = plt.gca()
    ax.set_ylim([-2.5, 2.5])
    plt.show()
```
